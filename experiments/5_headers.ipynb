# **Random User-Agent**

* `requests.get('https://www.ukaalo.com/') # <Response [403]>`

* but using a web browser user header like 'NX NIFM/00' gives correct response
`requests.get('https://www.ukaalo.com/', headers={'User-Agent':'NX NIFM/00'}) # <Response [200]>`

* [this ]('https://explore.whatismybrowser.com/useragents/explore/software_type_specific/web-browser/1') site contains a list of random headers. so crawling it to get a list of 500 headers used by different browsers

```python
from bs4 import BeautifulSoup

# Sample HTML content
html_content = '''
<html>
    <body>
        <table class="table">
            <tr><td>Row 1, Column 1</td><td>Row 1, Column 2</td></tr>
            <tr><td>Row 2, Column 1</td><td>Row 2, Column 2</td></tr>
            <tr><td>Row 3, Column 1</td><td>Row 3, Column 2</td></tr>
        </table>
    </body>
</html>
'''

# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Find the table with class "table"
table = soup.find('table', class_='table')

# Extract text from all <td> elements
data = [td.get_text(strip=True) for td in table.find_all('td')]

# Print the extracted text
print(data)
```
